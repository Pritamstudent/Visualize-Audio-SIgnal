# Visvualizing Audio Signals

### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
# Read the audio file
sampling_freq, signal = wavfile.read('/content/harvard.wav')
```

```python
print("\nSignal shape: ", signal.shape)
print("Datatype: ", signal.dtype)
print('Signal duration: ', round(signal.shape[0] / float(sampling_freq), 2), 'seconds')
```

Let's break down the steps:

The signal variable holds the original signal data.
np.power(2, 15) calculates 2 raised to the power of 15, which is equal to 32768.<br>

The signal is divided by 32768 to scale the values between -1 and 1 (assuming the original signal was represented with 16-bit signed integers).

```python
# Normalize the signal
signal = signal / np.power(2, 15)
```

```python
# Extract the first 50 values
signal = signal[:50]
```

```python
# Construct the time axis in milliseconds
time_axis = 1000 * np.arange(0, len(signal) , 1) / float(sampling_freq)
```

```python
# Plot the audio signal
plt.plot(time_axis, signal, color = 'r')
plt.xlabel('Time in Milliseconds')
plt.ylabel('Amplitude')
plt.title('Input audio signal')
plt.show()
```
